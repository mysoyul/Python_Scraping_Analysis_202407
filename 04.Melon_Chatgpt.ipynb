Melon 100 chart 에서 100곡의 노래 정보를 웹스크래핑 하는 source code를 작성해 주세요


```python
import requests
from bs4 import BeautifulSoup

# Melon 100 차트 URL
url = 'https://www.melon.com/chart/index.htm'

# 요청 헤더 설정 (Melon 사이트가 봇 접근을 막기 위해 설정)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Referer': 'https://www.melon.com/index.htm'
}

# 페이지 요청 및 파싱
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# 곡 정보가 담긴 리스트 가져오기
songs = soup.select('tr[data-song-no]')

# 각 곡의 정보 추출
song_data = []
for song in songs:
    song_no = song['data-song-no']
    title = song.select_one('div.ellipsis.rank01 > span > a').text
    artist = song.select_one('div.ellipsis.rank02 > a').text
    album = song.select_one('div.ellipsis.rank03 > a').text
    rank = song.select_one('span.rank').text

    song_data.append({
        'rank': rank,
        'song_no': song_no,
        'title': title,
        'artist': artist,
        'album': album
    })

# 결과 출력
for data in song_data:
    print(f"Rank: {data['rank']} | Title: {data['title']} | Artist: {data['artist']} | Album: {data['album']} | Song No: {data['song_no']}")

# 파일로 저장 (선택사항)
import csv

with open('melon_top_100.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['rank', 'song_no', 'title', 'artist', 'album'])
    writer.writeheader()
    writer.writerows(song_data)
```
